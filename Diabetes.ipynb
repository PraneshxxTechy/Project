**Importing pandas**

```python
import pandas as pd
```

**Load and Prepare Data**

```python
data = pd.read_csv('diabetes.csv')
```

**X** is the **feature-set** and **y**is the **target (or) label**. We extract X and y from the datset using **iloc** function from pandas library.In this dataset, **Outcome** column is taken as target leaving the others as **feature-set**.

```python
X = data.drop("Outcome", axis=1)
```

```python
y = data["Outcome"]
```

Importing **keras**. It is used to **build** to **neural network**.


```python
from keras.models import Sequential
from keras.layers import Dense
```

Defining the **NN** based on the **dataset**

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

**Compiling the network**

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

**Fitting the data in the model**

```python
model.fit(X,y, epochs=150, batch_size=10)
```

**Evaluation of performance of the model**

```python
_, accuracy = model.evaluate(X, y)
print("Model accuracy: %.2f"% (accuracy*100))
```

**Making Predictions**

```python
predictions = model.predict(X)
print(predictions)
```

We **round off** the values as it gives the **probability** of the classes. Here, we have **two classes(namely 0 and 1)**, so if the value is **above 5** it is **predicted as 1** else 0.

```python
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
