**Importing** the required **libraries**

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from IPython.display import display
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

Reading the **csv file** using **Pandas** library

```python
df=pd.read_excel(r"/content/Dry_Bean_Dataset.xlsx")
```

```python
display(df)
```

**Label Encoder** is used for transforming **string-type** classes to **integers**

```python
l=LabelEncoder()
df['Class']=l.fit_transform(df['Class'])
```

```python
display(df)
```

X is the **feature-set** and y is the **target (or) label**. We extract X and y from the datset using **iloc** function from pandas library.In this dataset, **Class** column is taken as target leaving the others as feature-set.

```python
X=df.iloc[:,:16]
print(X)
```

```python
y=df.iloc[:,16]
```

```python
X.isnull().sum()
```

**StandardScaler** to scale the dataset as the **feature** with **large value** would dominate the others. It works by making the **mean zero**.

```python
s=StandardScaler()
```

**fit_transform** helps in scaling the dataset with **StandardScaler()**

```python
X=pd.DataFrame(s.fit_transform(X),columns=['Area','Perimeter'	,'MajorAxisLength','MinorAxisLength','AspectRation','Eccentricity','ConvexArea','EquivDiameter','Extent','Solidity','roundness','Compactness','ShapeFactor1','ShapeFactor2','ShapeFactor3','ShapeFactor4'])
```

**Train_test_split** is used to split the dataset into two parts namely **Train** and **test**. The train part is used to for the **training purpose** of the model while the test part is used for the **evaluation of the model**.

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

**Model Selection**

```python
model1=RandomForestClassifier()
```

**Fitting** the **train data** into the model

```python
model1.fit(X_train,y_train)
```

**Making predictions**

```python
pred1=model1.predict(X_test)
```

**Evalution** of the **performance** of the model

```python
acc1=accuracy_score(pred1,y_test)
print(acc1)
```
